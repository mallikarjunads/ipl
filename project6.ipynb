```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
%matplotlib inline
```

```python
df = pd.read_csv("ipl.csv")
```

```python
print('Checking the index values with column 1',all(df.iloc[:,0].isin(df.index.values)))
print('Checking the index values with column 2',all(df.iloc[:,1].isin(df.index.values)))

df.head(3)
#shape of the dataset
print('Shape of the DataFrame: ',df.shape)

#getting the name of the first colmuns to drop them
print(df.iloc[:,0].name)
print(df.iloc[:,1].name)

print('columns: ',df.columns.values)
#droping the columns from axis 1
df.drop(['Unnamed: 0','Unnamed: 0.1'], axis=1,inplace=True)
```

```python
df.Date = df.Date.str.replace(',',' ')
df.Year = df.Year.apply(lambda x: str(x))
df.Date = df.Date + ' ' + df.Year

#ff = dff.Date.apply(lambda x:datetime.strptime(str(x),'%b%d%a'))
df.Date = pd.to_datetime(df.Date,format='%b %d %a %Y')
df.Tied = df.Tied.notnull().astype('int')

#changing the string values of won_runs into integers
df.won_runs = df.won_runs.str.replace(r'[^0-9]','', regex=True)
df.won_runs.fillna(0,inplace=True)
df.won_runs = df.won_runs.astype('int64')

#changing the string values of won_wickets into integers
df.won_wickets.fillna('0',inplace=True)
df.won_wickets = df.won_wickets.str.replace(' ','')
df.won_wickets = df.won_wickets.apply(lambda c:c[0])
df.won_wickets = df.won_wickets.astype('int64')
```

```python
df
```

```python
# short_name = [x for x in input().split(' ')]
# full_name = df.Team1.unique().tolist()


#changing the values of team name into their short forms
full_name = ['ROYAL CHALLENGERS BANGALORE',
 'KINGS XI PUNJAB',
 'DELHI CAPITALS',
 'KOLKATA KNIGHT RIDERS',
 'MUMBAI INDIANS',
 'RAJASTHAN ROYALS',
 'DECCAN CHARGERS',
 'CHENNAI SUPER KINGS',
 'KOCHI TUSKERS KERALA',
 'PUNE WARRIORS',
 'SUNRISERS HYDERABAD',
 'GUJARAT LIONS',
 'RISING PUNE SUPERGIANT','RISING PUNE SUPERGIANTS','DELHI DAREDEVILS','NO RESULT','HYDERABAD']
short_name = ['RCB',
 'KXP',
 'DELHI',
 'KKR',
 'MI',
 'RR',
 'DECCAN',
 'CSK',
 'KTK',
 'PW',
 'SH',
 'GL',
 'RPS','RPS','DELHI','MATCH ABANDONED','SH']

temp = {}
for key in full_name:
    for value in short_name:
        temp[key] = value
        short_name.remove(value)
        break
        

#mapping the dictionary values with Toss column        
toss_list = df.Toss.unique().tolist()
#toss_values = [x for x in input().split(' ')]
toss_values = ['RCB','CSK','RR','DECCAN','MI','KXP','KKR','DELHI','NOT_DECIDED','KTK','8PM','PW','SH','4PM','GL','RPS']

toss_temp = {}
for key in toss_list:
    for value in toss_values:
        toss_temp[key] = value
        toss_values.remove(value)
        break
```

```python

#mapping the dictionary values to Result column
df.Team1 = df.Team1.replace(temp,regex=True)
df.Team2 = df.Team2.replace(temp,regex=True)

df.Team1 = df.Team1.str.replace(' ','')
df.Team2 = df.Team2.str.replace(' ','')

df['Result'] = df['Result'].str.split(' won').str[0].str.upper()
df = df.replace({"Result": temp})

df['Result'] = df['Result'].str.split(' WITHOUT').str[0].str.upper()
df['Result'] = df['Result'].str.split(' \(').str[0].str.upper()
df['Result'] = df['Result'].str.split(' -').str[0].str.upper()

print('Unique values of Result column',df.Result.unique())
     
df.Toss = df.Toss.map(toss_temp)

print('Unique values of Toss column',df.Result.unique())
```

```python
df
```

```python
State = df.Place.str.split(', ').str[1]
State.name = 'State'
df = pd.concat([df,State],axis=1)
df.Place = df.Place.str.split(', ').str[0]
df.Place = df.Place.str.split(' Stadium').str[0]
#formatt = '%I:%M %p GMT / %I:%M %p LOCAL' 

df.drop(['Time'],axis=1,inplace=True)
# df['Time'].unique()
# df['Time'] = pd.to_datetime(df['Time'],format = formatt)
# df['Time'] = [time.time() for time in df['Time']]
```

```python
toss_des = {'bat':'1','bowl':'0','PM':'2'}
df.TossDecision = df.TossDecision.map(toss_des).astype('int64')
df.to_csv('df_cleaned.csv',index=False)
df.head()
```
